# **Introduction**

The dataset represents transactional data from a pizza delivery/restaurant service, capturing detailed information about customer orders. Each entry corresponds to an individual pizza item within an order, including specifics about the pizza, order timing, and financial details.

## Dataset Overview
The dataset contains information about pizza orders. The columns include:
- `pizza_id`: Unique identifier for the pizza.
- `order_id`: Unique order identifier.
- `pizza_name_id`: Identifier for the type of pizza.
- `quantity`: Number of pizzas in the order.
- `order_date`: Date the order was placed.
- `order_day`: Day of the week.
- `order_time`: Time the order was placed.
- `unit_price`: Price of a single pizza.
- `total_price`: Total price of the order.
- `pizza_size`: Size of the pizza (e.g., M, L).
- `pizza_category`: Type of pizza (e.g., Veggie, Classic).
- `pizza_ingredients`: Ingredients in the pizza.

# **Split data into training and testing:**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
```

# ***Feature Scaling:***

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

numerical_cols = x.select_dtypes(include=['float64', 'int64']).columns

x_train_num = x_train[numerical_cols]
x_test_num = x_test[numerical_cols]

scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train_num)
x_test_scaled = scaler.transform(x_test_num)

```

# **Build the Machine Learning Model:**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

regressor = {
    "Linear Regression": LinearRegression(),
    "Decision Tree Regressor": DecisionTreeRegressor(),
    "Random Forest Regressor": RandomForestRegressor(),
    "Support Vector Regressor": SVR(),
    "K-Nearest Neighbors Regressor": KNeighborsRegressor(),
}

for name, clf in regressor.items():
    clf.fit(x_train_scaled, y_train)
    y_pred = clf.predict(x_test_scaled)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"Regressor: {name}")
    print(f"Mean Squared Error: {mse}")
    print(f"R-squared: {r2}")
    print("-" * 20)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
plot_tree(regressor['Decision Tree Regressor'].fit(x_train_scaled, y_train), feature_names= x.columns, filled= True, rounded= True)
plt.show()
```

```python
result = []

for name, clf in regressor.items():
    clf.fit(x_train_scaled, y_train)
    y_pred = clf.predict(x_test_scaled)
    R2_score = r2_score(y_test, y_pred)
    result.append([name, R2_score])

result_df = pd.DataFrame(result, columns=['regressor', 'R2_score'])

best_classifier = result_df.loc[result_df['R2_score'].idxmax()]
print(f"The Best Regressor is: {best_classifier['regressor']} with R2 score: {best_classifier['R2_score']}")

best_clf_name = best_classifier['regressor']
for name, clf in regressor.items():
    if name == best_clf_name:
        best_clf = clf
        break
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = best_clf.predict(x_test_scaled)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
best_clf_name = best_classifier['regressor']
for name, clf in regressor.items():
    if name == best_clf_name:
        best_clf = clf
        break

y_pred = best_clf.predict(x_test_scaled)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

print("\nModel Performance Interpretation:")
print(f"The best performing model is {best_clf_name}.")

print(f"Mean Squared Error (MSE): {mse}")
print("MSE represents the average squared difference between the predicted and actual values.")
print("A lower MSE indicates better model performance.")

print(f"R-squared (R2): {r2}")
print("R2 represents the proportion of variance in the dependent variable (total_price) that is predictable from the independent variable (unit_price).")
print("An R2 value closer to 1 indicates a better fit, while a value closer to 0 suggests a poor fit.")

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.show()
```
